```python
#correlation heatmap and skewness report
import numpy as np

# correlation heatmap 
numeric_colls = data.select_dtypes(include=[np.number]).columns
corr_matrix = data[numeric_colls].corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()

#finding highyl correlated pairs (>0.8 or <-0.8)
# Find highly correlated pairs (>0.8 or <-0.8), excluding self-correlations
high_corr_pairs = []
threshold = 0.8
for col1 in corr_matrix.columns:
    for col2 in corr_matrix.columns:
        if col1 != col2:
            corr_val = corr_matrix.loc[col1, col2]
            if abs(corr_val) > threshold:
                sorted_pair = tuple(sorted([col1, col2]))
                if sorted_pair not in high_corr_pairs:
                    high_corr_pairs.append(sorted_pair)

print("Highly correlated pairs (>|0.8|):")
for pair in high_corr_pairs:
    print(f"{pair[0]} & {pair[1]}: {corr_matrix.loc[pair[0], pair[1]]:.2f}")

```

```python
# Investigate zero/negative values in bill and payment columns

bill_cols = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']
pay_cols = ['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']

# Check negative values in bill amount columns
print("Negative values in bill columns:")
for col in bill_cols:
    neg_count = (data[col] < 0).sum()
    print(f"{col}: {neg_count}")

# Check zero values in bill amount columns
print("\nZero values in bill columns:")
for col in bill_cols:
    zero_count = (data[col] == 0).sum()
    print(f"{col}: {zero_count}")

# Check negative values in payment columns (should not be negative)
print("\nNegative values in payment columns:")
for col in pay_cols:
    neg_count = (data[col] < 0).sum()
    print(f"{col}: {neg_count}")

# Check zero values in payment columns
print("\nZero values in payment columns:")
for col in pay_cols:
    zero_count = (data[col] == 0).sum()
    print(f"{col}: {zero_count}")

# Flag impossible values in bills (e.g., extremely negative amounts)
impossible_bills = {}
for col in bill_cols:
    # Flag rows with bill amount < -100 or an arbitrary large negative value
    impossible_rows = data[data[col] < -100]
    impossible_bills[col] = impossible_rows.shape[0]
print("\nFlagged potentially impossible negative values (less than -100) in bill columns:")
for col, count in impossible_bills.items():
    print(f"{col}: {count}")

# Flag impossible values in payments (e.g., negative payment, unusually large outliers)
impossible_payments = {}
for col in pay_cols:
    # Flag rows with payment < 0 or payment > 1,000,000 as impossible
    impossible_rows = data[(data[col] < 0) | (data[col] > 1e6)]
    impossible_payments[col] = impossible_rows.shape[0]
print("\nFlagged potentially impossible payment values (negative or above 1,000,000):")
for col, count in impossible_payments.items():
    print(f"{col}: {count}")

```

```python
# helpers and reshaped frmaes
import numpy as np 

bill_cols = [f'BILL_AMT{i}' for i in range(1, 7)]
pay_amt_cols = [f'PAY_AMT{i}' for i in range(1, 7)]
pay_status_cols = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']

bills = data[bill_cols].copy()
pays = data[pay_amt_cols].copy()
pay_status = data[pay_status_cols].copy()

# Visualise repayment trajectories 
sample_isd = np.random.choice(data.index, size=6, replace=False)

fig, axes = plt.subplots(2,3, figsize=(12,12))
axes = axes.flatten()

for ax, idx in zip(axes, sample_isd):
    ax.plot(range(6), bills.loc[idx], marker='o', label='Bill amount')
    ax.plot(range(6), pays.loc[idx], marker='x', label='Payment amount')
    ax.set_title(f'Customer {idx}, target={data.loc[idx, target_col]}')
    ax.set_xticks(range(6))
    ax.set_xticklabels([f'M{i}' for i in range(1, 7)])
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
bill_deltas = bills.diff(axis=1).fillna(0)
pay_deltas = pays.diff(axis=1).fillna(0)

data['bill_slope'] = bill_deltas.mean(axis=1)
data['bill_volatility'] = bills.std(axis=1)
data['pay_slope'] = pay_deltas.mean(axis=1)
data['pay_volatility'] = pays.std(axis=1)

pay_to_bill = (pays / bills.replace(0, np.nan)).clip(0, 5).fillna(0)
data['pay_to_bill_mean'] = pay_to_bill.mean(axis=1)
data['pay_to_bill_min'] = pay_to_bill.min(axis=1)
data['pay_to_bill_last'] = pay_to_bill.iloc[:, -1]
```

```python
late_flags = (pay_status > 0).astype(int)
severe_flags = (pay_status >= 2).astype(int)

data['late_months'] = late_flags.sum(axis=1)
data['severe_late_months'] = severe_flags.sum(axis=1)
data['recent_late'] = late_flags.iloc[:, -1]

def longest_streak(row):
    streak = best = 0
    for val in row:
        if val:
            streak += 1
            best = max(best, streak)
        else:
            streak = 0
    return best

data['late_streak'] = late_flags.apply(longest_streak, axis=1)
data['severe_streak'] = severe_flags.apply(longest_streak, axis=1)
```

```python
first_half = slice(0, 3)
second_half = slice(3, 6)

data['bill_mean_early'] = bills.iloc[:, first_half].mean(axis=1)
data['bill_mean_late'] = bills.iloc[:, second_half].mean(axis=1)
data['bill_growth'] = data['bill_mean_late'] - data['bill_mean_early']

data['pay_mean_early'] = pays.iloc[:, first_half].mean(axis=1)
data['pay_mean_late'] = pays.iloc[:, second_half].mean(axis=1)
data['pay_growth'] = data['pay_mean_late'] - data['pay_mean_early']
```

```python
temporal_cols = [
    'bill_slope', 'bill_volatility', 'pay_slope', 'pay_volatility',
    'pay_to_bill_mean', 'pay_to_bill_min', 'pay_to_bill_last',
    'late_months', 'severe_late_months', 'recent_late',
    'late_streak', 'severe_streak',
    'bill_growth', 'pay_growth'
]

grouped = data.groupby(target_col)[temporal_cols].mean().T
display(grouped)

plt.figure(figsize=(8, 10))
sns.barplot(
    x=grouped[1] - grouped[0],
    y=grouped.index,
    orient='h',
    palette='coolwarm'
)
plt.title('Mean difference (default - non-default)')
plt.xlabel('Difference')
plt.show()
```

```python
# === 3) Multicollinearity probe ===
from statsmodels.stats.outliers_influence import variance_inflation_factor

numeric_cols = data.select_dtypes(include=[np.number]).drop(columns=['ID']).columns
X = data[numeric_cols]

vif = pd.DataFrame({
    'feature': numeric_cols,
    'VIF': [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
}).sort_values('VIF', ascending=False)

display(vif.head(15))

```

```python
# === 4) Outlier inspection & scaling decisions ===
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(x=data['LIMIT_BAL'], ax=axes[0])
sns.histplot(np.log1p(data['PAY_AMT1']), bins=40, ax=axes[1])
axes[0].set_title('LIMIT_BAL Boxplot')
axes[1].set_title('PAY_AMT1 log distribution')
plt.tight_layout()
plt.show()

# Conclusion: payment/bill amounts are extremely skewed, so log1p scaling
# and clipping high percentiles should stabilize training.
```

```python
# === 5) Baseline predictive check ===
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

temporal_features = [
    'bill_slope', 'bill_volatility', 'pay_slope', 'pay_volatility',
    'pay_to_bill_mean', 'pay_to_bill_min', 'pay_to_bill_last',
    'late_months', 'severe_late_months', 'recent_late',
    'late_streak', 'severe_streak', 'bill_growth', 'pay_growth'
]

baseline_feats = ['LIMIT_BAL', 'AGE'] + temporal_features
X_train, X_test, y_train, y_test = train_test_split(
    data[baseline_feats], data[target_col],
    test_size=0.2, stratify=data[target_col], random_state=42
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

clf = LogisticRegression(max_iter=200, class_weight='balanced')
clf.fit(X_train_scaled, y_train)
auc = roc_auc_score(y_test, clf.predict_proba(X_test_scaled)[:, 1])
print(f"Baseline logistic AUC with temporal features: {auc:.3f}")

# This quick check confirms temporal features carry signal before we
# invest in fancier models. We'll replicate the exact transformations
# in src/preprocess.py and plug a more robust estimator in src/model.py.
```
