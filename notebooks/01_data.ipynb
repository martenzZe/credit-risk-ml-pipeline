```python
#check for missing values and duplicates 
print("Missing values per column:\n", data.isnull().sum().sort_values(ascending=False))
print("\nNumber of duplicate rows:", data.duplicated().sum())
```

```python
#investigate target variable
target_col = 'default.payment.next.month'
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x=target_col, data=data)
plt.title('Distribution of Target Variable')
plt.show()
```

```python
#categorical feature distributions
categorical_columns = ['SEX', 'EDUCATION', 'MARRIAGE']


for col in categorical_columns:
    print(col, data[col].value_counts())
    display(data.groupby(col)[target_col].agg(['count','mean']).sort_values('mean', ascending=False))
```

### Deep Numeric Feature analysis

```python
# boxplots for Limit_BAL, AGE, top BILL_AMT, top PAY_AMT split by target
key_numerics = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'BILL_AMT2', 'PAY_AMT1', 'PAY_AMT2']

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, col in enumerate(key_numerics):
    sns.boxplot(x=target_col, y=col, data=data, ax=axes[i])
    axes[i].set_title(f'Boxplot of {col} by {target_col}')
plt.tight_layout()
plt.show()

# grouped statistics 
print("\nMean values by target:")
grouped_means=data.groupby(target_col)[key_numerics].mean()
print(grouped_means)
```

```python
#correlation heatmap and skewness report
import numpy as np

# correlation heatmap 
numeric_colls = data.select_dtypes(include=[np.number]).columns
corr_matrix = data[numeric_colls].corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()

#finding highyl correlated pairs (>0.8 or <-0.8)
```
